```python
import requests
```

```python
URL = 'https://app.rakuten.co.jp/services/api/Travel/KeywordHotelSearch/20170426'
APP_ID = '1099529787563736246'
```

```python
params = {
    'applicationId' : APP_ID,
    'keyword' : '沖縄'
}
```

```python
res = requests.get(URL, params)
```

```python
res.status_code
```

```python
hotels = res.json()['hotels']
hotels
```
